```python
import cv2
```

```python
tracker=cv2.TrackerBoosting_create()
def ask_for_tracker():
    print("CHOOSE API FOR USE")
    print("0 FOR BOOSTING")
    print("1 FOR MIL")
    print("2 FOR KCF")
    print("3 FOR TLD")
    print("4 FOR MEDIANFLOW")
    choice =input("SELECT YOUR TRACKER:")
    if choice=="0":
        tracker=cv2.TrackerBoosting_create() #very well known but does not know when tracking fail
    if choice=="1":
        tracker=cv2.TrackerMIL_create() # good performance but cant recover from full obstruction
    if choice=="2":
        tracker=cv2.TrackerKCF_create() #better than boosting and mil but cant recover from full obstruction
    if choice=="3":
        tracker=cv2.TrackerTLD_create() # can handle large change good with obstruction but can provide many false positivies
    if choice=="4":
        tracker=cv2.TrackerMedianFlow_create() #very good at reporting fail and works well with predictable motion  but fails under large motion

    return tracker
```

```python
tracker=ask_for_tracker()
tracker_name=str(tracker).split()[0][1:]
cap=cv2.VideoCapture(0)
ret, frame= cap.read()
roi=cv2.selectROI(frame,False)

ret=tracker.init(frame,roi)

while True:
    ret, frame= cap.read()

    success, roi= tracker.update(frame)
    (x,y,w,h)=tuple(map(int,roi))
    if success:
        p1=(x,y)
        p2=(x+w,y+h)
        cv2.rectangle(frame,p1,p2,(0,255,0),3)
    else:
        cv2.putText(frame, "Failure to Detect Tracking!!", (100,200), cv2.FONT_HERSHEY_SIMPLEX, 1,(0,0,255),3)
        
    cv2.imshow(tracker_name,frame)
    k=cv2.waitKey(1) & 0xFF
    if k== 27:
        break
    
  

cap.release()
cv2.destroyAllWindows()

        
        
    

```
